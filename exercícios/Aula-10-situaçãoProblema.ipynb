```python
caminho = ('../ArquivosAnálise/IceCreamData.csv')

dados  = pd.read_csv(caminho)

dados.head(1)
```

#### RENOMEANDO AS COLUNAS

```python
dados = dados.rename(columns={'Temperature': 'Temperatura','Revenue': 'Receitas'})
```

```python
# Final do dataframe
dados.tail(1)
```

```python
# Descrição do datafreme
dados.describe()
```

```python
# Informação do datafreme
dados.info()
```

#### VIZUALIZAR OS DADOS
- Etapa 03

```python
# Gráfico para analizar a correlação e distribuição dos dados

sns.jointplot(data=dados, x='Temperatura', y= 'Receitas', color='darkblue')
plt.savefig('../graficos/Temperatura x Receitas.png')
```

```python
# Gráfico para analizar a correlação e distribuição dos dados
sns.pairplot(dados)
plt.savefig('../graficos/Pair Plot - Temperatura x Receitas.png')
```

```python
# Gráfico para analizar a correlação e distribuição dos dados

sns.lmplot(data= dados, x ='Temperatura', y = 'Receitas')
plt.savefig('../graficos/Gráfico de Dispersão - Temperatura x Receitas.png')
```
