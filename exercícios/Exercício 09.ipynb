```python
import pandas as pd # Manipular e tratar dados
import matplotlib.pyplot as plt # Gerar gráficos
import os # Gerenciar o sistema operacional

caminho = '../ArquivosAnálise/company_sales_data.csv'

dados = pd.read_csv(caminho)

dados.head()
```

### Grafico-01: Month_number x total_profit (plot)

```python
# CONVERTENDO AS COLUNAS EM LISTA
listaMeses = dados['month_number'].to_list()
listaLucro = dados['total_profit'].to_list()

# CRIAR OS DADOS EM FUNÇÃO DOS DADOS ANTERIORES
plt.plot( listaMeses,  # EIXO X
          listaLucro,  # EIXO Y
          label = 'LUCRO', #RÓTULO DO GRÁFICO
          color ='#680DE0', # COR DA LINHA
          lw = 3,           # ESPESSURA DA LINHA  
          ls = '--',        # ESTILO DA LINHA
          marker = 'o'      # MARCADOR DO GRÁFICO
          )


plt.title('LUCRO POR MÊS', fontsize = 18, c= '#2C2E33')

plt.xlabel('MESES', fontsize = 10,)
plt.ylabel('LUCRO', fontsize = 10,)

plt.xticks(fontsize = 8, c= 'darkblue', fontweight= 'bold')
plt.yticks(fontsize = 8, c= 'darkblue', fontweight= 'bold')

plt.legend()

#os.mkdir('../graficos/grafico-imagens')
plt.savefig('../graficos/grafico-imagens/grafico_01.png')
```

#### Exercício 03 – Leia todos os dados de vendas de produtos e mostre-os usando um 
gráfico multilinha. 
Exiba o número de unidades vendidas por mês para cada produto usando gráficos 
multilinhas. (ou seja, faça a plotagem das linhas separadas para cada produto).

```python
# CONVERTER COLUNAS EM LISTAS

listaMeses = dados['month_number'].to_list() # EIXO X
listaCremeDental = dados['toothpaste'].to_list()
listaLimpaFacial = dados['facewash'].to_list()
listaFacial = dados['facecream'].to_list()
listaSabonete = dados['bathingsoap'].to_list()
listaShampoo = dados['shampoo'].to_list()
listaHidratante = dados['moisturizer'].to_list()

# PLOTAR OS DADOS NO GRÁFICO
plt.plot(listaMeses, listaCremeDental, label ='Creme Dental', marker = 'o', color = '#57FABA', ls = '--')
plt.plot(listaMeses, listaLimpaFacial, label = 'Limpador Facial', marker = 'o', color = '#9F57FA', ls = '--')
plt.plot(listaMeses, listaFacial, label = 'Creme Facial', marker = 'o', color = '#57FAF1', ls = '--')
plt.plot(listaMeses, listaSabonete, label ='Sabonete', marker = 'o', color = '#53FA44', ls = '--')
plt.plot(listaMeses, listaShampoo, label = 'Shampoo', marker = 'o', color = '#FAF751', ls = '--')
plt.plot(listaMeses, listaHidratante, label = 'Hidratante Corporal', marker = 'o', color = '#FA9C57', ls = '--')

# ADICIONA TÍTULO
plt.title('HISTÓRICO DE VENDAS', fontsize = 15, color = '#0FBFFA')
# ADIIONA TÍTULO NO EIXO X
plt.xlabel('MESES', fontsize = 15, color = '#0FBFFA')
# ADIIONA TÍTULO NO EIXO Y
plt.ylabel('LUCRO', fontsize = 15, color = '#0FBFFA')

plt.legend(fontsize = 8)
# SALVANDO A IMAGEM EM UMA PASTA
plt.savefig('../graficos/grafico-imagens/exercicio03.png')
```

#### Exercício 04 
Leia os dados de vendas de pasta de dente de cada mês e mostre-os usando um gráfico de dispersão. Além disso, adicione uma grade no gráfico. o estilo da linha de grade deve “–“. 
O gráfico de dispersão deve ficar assim

```python
listaMeses = dados['month_number'].to_list()
listaVendasCremeDental = dados['toothpaste'].to_list()
# PLOTANDO COM GRÁFICO DE DISPERSÃO
plt.scatter(listaMeses, listaVendasCremeDental, label = 'Creme Dental Vendas')

plt.legend()

# ADICIONANDO LINHAS DE GRADE AO GRÁFICO
plt.grid(True, linewidth = 0.5, linestyle = '--', color = '#9F57FA')

plt.title('TOOTH PASTE SALES DATA', fontsize = 10)

plt.xlabel('MESES')
plt.ylabel('UNIDADES VENDIDAS')
```

#### Exercício 05
Leia os dados de vendas de cremes faciais e produtos de lavagem facial e mostre-os usando o gráfico de barras. O gráfico de barras deve exibir o número de unidades vendidas por mês para cada produto. Adicione uma barra separada para cada produto no mesmo gráfico. O gráfico de barras deve ficar assim.

```python
listaMeses = dados['month_number'].to_list()
listaVendasCremeDental = dados['toothpaste'].to_list()
listaLimpaFacial = dados['facewash'].to_list()

# CRIAR UM GRÁFICO DE COLUNAS 

plt.bar([pos-0.25 for pos in listaMeses], listaVendasCremeDental, label ='Creme Dental', width = 0.25)
plt.bar([pos+0.25 for pos in listaMeses], listaLimpaFacial, label = 'Creme Facial', width = 0.25 )
plt.legend()

plt.title('VENDAS PRODUTOS')
plt.xlabel('MESES')
plt.ylabel('vendas')
```
