#### REPRESENTAÇÃO GRÁFICA COM BOX PLOT

```python
# ANÁLISE DE OUTLIERS
fig, axs = plt.subplots(3, figsize = (5,5))
plt1 = sns.boxplot(vendas['TV'], ax = axs[0], orient='h')
plt2 = sns.boxplot(vendas['Newspaper'], ax = axs[1], orient='h')
plt3 = sns.boxplot(vendas['Radio'], ax = axs[2], orient='h')

plt.tight_layout() # PERMITIR MOSTRAR O RÓTULO PARA CADA VARIÁVEL
```

#### GRÁFICO DE DISPERSÃO

```python
plt.scatter(data=vendas,x= 'TV', y='Sales', color ='darkblue')
plt.title('Análise de correlação TV X Sales')
plt.xlabel('Investimentos em TV')
plt.ylabel('SALES')
plt.grid(True)
```

```python
plt.scatter(data=vendas,x= 'Newspaper', y='Sales', color ='darkblue')
plt.title('Análise de correlação Newspaper X Sales')
plt.xlabel('Investimentos em Newspapper')
plt.ylabel('SALES')
plt.grid(True)
```

```python
plt.scatter(data=vendas,x= 'Radio', y='Sales', color ='darkblue')
plt.title('Análise de correlação Radio X Sales')
plt.xlabel('Investimentos em Radio')
plt.ylabel('SALES')
plt.grid(True)
```

#### GRÁFICO DE DISPERSSÃO COM PAIRPLOT

```python
sns.pairplot(
    vendas,
    x_vars=['TV','Newspaper','Radio'],
    y_vars='Sales',
    height=4,
    aspect=1,
    kind='scatter'
)
```
