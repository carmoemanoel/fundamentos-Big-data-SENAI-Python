#### MATPLOTLIB – PRIMEIROS PASSOS
Matplotlib é a principal biblioteca para geração de gráficos em Python. A
biblioteca possui uma infinidade de possibilidades, sendo possível criarmos desde
gráficos simples até outros mais elaborados.


#### IMPORTANDO A BIBLIOTECA & CRIANDO O PRIMEIRO GRÁFICO

```python
# Importando as bbliotecas
import matplotlib.pyplot as plt
import numpy as np

# Criar os arrays (matrizes) com valores

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Criar o gráfico

plt.plot(x,y)
```

#### Modificando o tipo de marcador

```python
plt.plot(x, y, marker = '*', c = 'darkblue')
```

#### Mudando como os pontos são ligados e o tipo de linha

```python
plt.plot(x,y,marker='x', c='darkblue',ds='steps',ls='--', lw=3)
```

#### Alterar a cor dos marcadores

```python
plt.plot(x, 
         y, 
         marker='o', 
         c='darkblue', # cor da linha
         ls='--',      # line Style
         lw=2,         # Espessura da linha
         mec='red',    # Cor borda do marcador
         markerfacecolor = 'red' # ou MFC cor interna da linha 
         )   


# Adicionando títulos dos eixos

plt.xlabel('VALORES DE X', fontsize = 15)
plt.ylabel('VALORES DE Y', fontsize = 15)

# Adicionando título ao gráfico

plt.title('TÍTULO DO GRÁFICO', fontsize = 20, c='purple')

# Alterando os rótulos dos eixos

plt.xticks(fontsize = 8, c='blue')
plt.yticks(fontsize = 8, c='orange')
```
